## 📊 Column Descriptions

| Feature                        | Description |
|-------------------------------|-------------|
| **Gender**                    | Biological sex of the individual (Male/Female). |
| **Age**                       | Age in years. |
| **Height**                    | Height in meters (m). |
| **Weight**                    | Weight in kilograms (kg). |

### Data Transformation

for Gender column let's convert male to Male

```python
data['Gender'] = data['Gender'].replace("male", "Male")
```

```python
data['Gender'].unique()
```

for FAVC column we need to convert nope to no

```python
data['FAVC'] = np.where(data['FAVC'] == "Nope", "no", data['FAVC'])

data['FAVC'].unique()
```

For Family_history column we need to convert nope to no

```python
data['family_history_with_overweight'] = data['family_history_with_overweight'].replace("Nope", "no")

data['family_history_with_overweight'].unique()
```

For Smoke and CAEC Column we again need to convert no to nope

```python
data['CAEC'] = np.where(data["CAEC"] == "no", "Nope", data['CAEC'])

data['SMOKE'] = np.where(data["SMOKE"] == "no", "Nope", data['SMOKE'])
```

```python
data['CAEC'].unique()

```

```python
data['SMOKE'].unique()
```

For FCVC column (frequency of vegetable consumption) we need to have a data on Scale of 1 to 3 but we have continous data as it may appear hard for user to give input in that way so grouped these values in 1,2 and 3 

```python
my.minmax(data['FCVC'])
```

```python
data['FCVC'] = data['FCVC'].apply(lambda x : 1 if x <= 1 else 2 if x <= 2 and x > 1 else 3)
```

```python
data['FCVC'].unique()
```

For NCP COlumn i.e, number of Main Meals Perday we need again a scaled data not in a continous type as it is not user friendly 

```python
my.minmax(data['NCP'])
```

```python
data['NCP'] = data['NCP'].apply(lambda x : 1 if x <= 1 else 2 if x <= 2 and x > 1 else 3 if x <= 3 and x > 2 else 4)

data['NCP'].unique()
```

For CH2O column i.e, Concumption of water daily basis again a continous COlumn and we need to make it categorical and user friendly input

```python
my.minmax(data['CH2O'])
```

```python
data['CH2O'] = data['CH2O'].apply(lambda x : 1 if x <= 1 else 2 if x <= 2 and x > 1 else 3 )

data['CH2O'].unique()
```

For SCC Column we need to transform Nope values to no 

```python
data['SCC'] = np.where(data['SCC'] == 'Nope', "no", data['SCC'])
```

```python
data['SCC'].unique()
```

For FAF column i.e, Frequency of Physical Activity we need to have a scaled data so we are going to transform the data from continous nature to categorical Nature as it is always more user friendly

```python
my.minmax(data['FAF'])
```

```python
data['FAF'] = data['FAF'].apply(lambda x : 0 if x <= 0 else 1 if x > 0  and x <= 1 else 2 if x <= 2 and x > 1 else 3)

data['FAF'].unique()
```

for column name TUE that tells us about time spent on techinology so for that matter we need to categorise it let's look at the values so that we will categorise into low , high and medium usage

```python
data['TUE'] = data['TUE'].apply(lambda x : "Low Usage" if 0 <= x <= 0.7 else "Moderate Usage" if 0.7 < x <= 1.4 else "High Usage")

data['TUE'].unique()
```

#### For CAT-CAT Columns

```python
for col in cat_cols:
    cross_tab = pd.crosstab(df[col], df['health_condition'], normalize='index')

    # Plot
    ax = cross_tab.plot(kind='bar', stacked=True,
                        figsize=(10, 6),
                        color=['pink', 'lightgreen', 'lightblue', 'lavender', 'peachpuff'])

    # Title and labels
    plt.title(f'Health Condition by {col.replace("_", " ").title()}', fontsize=14, fontweight='bold')
    plt.xlabel(col.replace("_", " ").title(), fontsize=12)
    plt.ylabel('Proportion', fontsize=12)

    # Legend positioning
    plt.legend(title='Health Condition', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Rotate x labels if needed
    plt.xticks(rotation=20)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix, cohen_kappa_score, classification_report,
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve
)
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris # For demonstration

# --- Global variables to store model scores ---
# ये लिस्ट्स और DataFrame फ़ंक्शन के बाहर बनी हैं ताकि इनका डेटा हर बार सेव हो सके।
m = []
acc = []
pre = []
rec = []
f1 = []
kap = []
scoreline = pd.DataFrame()

def multiclass_model_validation(model, xtrain, ytrain, xtest, ytest):
    """
    Trains and evaluates a multi-class classification model.
    Asks the user if they want to save the performance scores.

    Parameters:
    model: The machine learning model to be trained and evaluated (e.g., RandomForestClassifier).
    xtrain (pd.DataFrame or np.array): Training features.
    ytrain (pd.Series or np.array): Training labels.
    xtest (pd.DataFrame or np.array): Testing features.
    ytest (pd.Series or np.array): Testing labels.
    """
    global scoreline, m, acc, pre, rec, f1, kap # इन सभी ग्लोबल वेरिएबल्स को फ़ंक्शन के अंदर इस्तेमाल करने के लिए घोषित किया गया है।

    print(f"\n🧠 Training: {model.__class__.__name__}")
    model.fit(xtrain, ytrain)

    # Probability predictions (for ROC-AUC)
    soft_pred = None
    try:
        soft_pred = model.predict_proba(xtest)
    except AttributeError:
        print("⚠️ This model does not support probability prediction. Skipping ROC-AUC.")

    # Hard predictions (for other metrics)
    hard_pred = model.predict(xtest)

    # --- Print Evaluation Metrics ---
    print('\n📊 Confusion Matrix:\n', confusion_matrix(ytest, hard_pred))
    print('🧮 Cohen Kappa Score:', round(cohen_kappa_score(ytest, hard_pred), 2))
    print('📋 Classification Report:\n', classification_report(ytest, hard_pred))

    # --- Plot ROC-AUC if possible ---
    if soft_pred is not None:
        try:
            classes = np.unique(ytest)
            # Binary target needed for multi-class ROC-AUC
            ytest_bin = label_binarize(ytest, classes=classes)

            # ROC-AUC plot only makes sense for more than one class
            if ytest_bin.shape[1] > 1:
                roc_auc = roc_auc_score(ytest_bin, soft_pred, average='macro', multi_class='ovr')
                print(f'📈 Macro ROC-AUC Score: {roc_auc:.4f}')

                plt.figure(figsize=(8, 6))
                for i in range(len(classes)):
                    fpr, tpr, _ = roc_curve(ytest_bin[:, i], soft_pred[:, i])
                    plt.plot(fpr, tpr, label=f'Class {classes[i]} (AUC={roc_auc:.2f})') # AUC score added to label for clarity
                plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
                plt.xlabel('False Positive Rate')
                plt.ylabel('True Positive Rate')
                plt.title('ROC Curves (One-vs-Rest)')
                plt.legend()
                plt.grid(True)
                plt.tight_layout()
                plt.show()
            else:
                print("⚠️ ROC-AUC can't be plotted for a single class in the test set.")
        except Exception as e:
            print(f"❌ ROC-AUC plot error: {e}")

    # --- Ask to save scores ---
    response = input("📌 Do you want to save this model's scores? (y/n): ")
    if response.strip().lower() == 'y':
        m.append(str(model.__class__.__name__)) # model object की जगह सिर्फ उसका नाम सेव करें
        acc.append(accuracy_score(ytest, hard_pred))
        # precision, recall, f1 में zero_division=0 जोड़ा गया है ताकि कोई डिवीज़न-बाय-ज़ीरो एरर न आए
        pre.append(precision_score(ytest, hard_pred, average='macro', zero_division=0))
        rec.append(recall_score(ytest, hard_pred, average='macro', zero_division=0))
        f1.append(f1_score(ytest, hard_pred, average='macro', zero_division=0))
        kap.append(cohen_kappa_score(ytest, hard_pred))

        # Update the global scoreline DataFrame
        scoreline = pd.DataFrame({
            'MODEL': m,
            'ACCURACY': acc,
            'PRECISION': pre,
            'RECALL': rec,
            'F1 SCORE': f1,
            'COHEN KAPPA SCORE': kap
        })
        print("✅ Scores saved to `scoreline`.")
    else:
        print("🚫 Scores not saved.")
```

RANDOM FOREST IS LOOKING BEST OF THE BEST BUT IT ALSO DOUBTFUL AS IT IS MORE PRONE TO OVERFITTING IT MIGHT BE THAT CASE SO TO CROSS CHECK THAT WE WILL GO WITH CROSS VALIDATION SCORE WITH DIFFERENT NUMBER OF CV

```python
from sklearn.model_selection import cross_val_score
import pandas as pd

def evaluate_models_cv(model, x, y, cv_list=[5, 10], scoring='f1_macro'):
    """
    Evaluates a classifier on multiple CV folds using specified scoring.
    Returns a DataFrame with mean and individual scores.

    Parameters:
    - model   : Scikit-learn compatible classifier
    - x, y    : Features and target
    - cv_list : List of integers for cross-validation folds
    - scoring : Metric like 'f1_macro', 'accuracy', etc.
    """
    results = []

    for cv in cv_list:
        scores = cross_val_score(model, x, y, cv=cv, scoring=scoring)
        results.append({
            'MODEL': model.__class__.__name__,
            'CV FOLDS': cv,
            'MEAN SCORE': round(scores.mean(), 4),
            'ALL SCORES': scores
        })

    return pd.DataFrame(results)

```

```python
li = [5, 7, 10, 15]
df_results = evaluate_models_cv(RandomForestClassifier(), x, y, cv_list=li, scoring='f1_macro')
df_results
```

This table shows how much each factor (like age, weight, or diet) influences or predicts a person's health condition. The "SCORE" and "PERCENTAGE" tell us how important each factor is. A higher score/percentage means the factor is more influential.

- 📈 **Inferences of Factors on Health Condition**

    - **Weight (45.7%)** is by far the **most important factor** in determining a person's health condition. This makes sense, as health conditions like "Overweight" or "Underweight" are directly defined by weight categories.
    - **Height (9.9%)** and **Age (8.8%)** are the **next most influential factors**, though much less important than weight. Height and age both play a role in health and weight assessments.
    -  **Snacking Frequency (5.4%)** also shows a noticeable influence.
    - **Number of Meals (3.9%)**, **Alcohol Consumption (3.0%)**, **Water Consumption (2.8%)**, **Physical Activity (2.8%)**, **Frequency of Vegetable Consumption (2.7%)**, and **Mode of Transportation (2.7%)** all have a moderate, somewhat similar level of influence.
    - **Family History with Overweight (2.5%)**, **Time Spent on Tech (2.1%)**, **Fried Food Consumption (1.2%)**, and **Calorie Monitoring (1.1%)** have a relatively smaller influence.
    - **Smoking (0.5%)** appears to be the **least influential factor** in this analysis for predicting health condition, although it's crucial to remember that smoking has many other significant negative health impacts not captured by this specific score.

        
**Takeaway:** While many factors play a role in health conditions, **Weight** stands out as the primary determinant. Other factors like Height, Age, Snacking Frequency, and daily habits (meals, physical activity, alcohol) also contribute, but to a lesser extent in this particular analysis.
